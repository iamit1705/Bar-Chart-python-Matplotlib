```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import style
```

```python
style.use('ggplot')

# Data for the plots
x = [2, 4, 6]
y = [12, 14, 16]

x2 = [3, 3, 4]
y2 = [7, 14, 5]

# Create bar plots with corrected alignment
plt.bar(x, y, color='r', align='center')
plt.bar(x2, y2, color='b', align='center')

# Add titles and labels
plt.title('info')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```
